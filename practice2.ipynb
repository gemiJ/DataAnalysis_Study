```scala
var movieRating = parsed3.filter(parsed3("_c1") === "1")
```

```scala
println("movieID : "+ "1")
movieRatingSummary.select("summary","_c2").show()
```

```scala
//각 영화 평점 평균,분산,최대,최소 값
for(movieid <- 1 to 1682){
    var movieRating = parsed3.filter(parsed3("_c1") === movieid)
    var movieRatingSummary =  movieRating.describe()
    println("movieID : "+ movieid)
    movieRatingSummary.select("summary","_c2").show()
}
```
